```python
%matplotlib qt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import random
x_o=np.arange(0,33,1)
y_o=np.arange(0,33,1)

plt.rcParams['backend'] = 'TkAgg'
plt.rcParams["figure.figsize"] = [8,8]
plt.rcParams["figure.autolayout"] = True
plt.rcParams['axes.axisbelow'] = True

X= random.choice(x_o)
Y= random.choice(y_o)

x_ul=33
x_ll=-1
y_ul=33
y_ll=-1

# Function to print mouse click event coordinates
def onclick(event):  
    x_c=event.xdata
    y_c=event.ydata
    x_clear=[-1,33,33,-1]
    y_clear=[-1,-1,33,33]
    global X
    global Y
    global x_ul
    global x_ll
    global y_ul
    global y_ll
    global ax
    global fig
    #print ([x_ll,x_ul,y_ll,y_ul])
    if x_c <=X and x_c >= X-1:
        if y_c <= Y and y_c >= Y-1:
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x=[X-1,X-1,X,X]
            y=[Y-1,Y,Y,Y-1]
            plt.fill(x,y,'blue')
            #print ([x_ll,x_ul,y_ll,y_ul])
            return ['Found',X,Y,x_c,y_c]
        elif int(y_c) < Y:
            y_ll=int(y_c)
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c), int(x_c)+1, int(x_c)+1, int(x_c)]
            y = [int(y_c), int(y_c), y_ul, y_ul]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['correct','UP',x_c,y_c])
            return ['correct','UP',x_c,y_c]
        else:
            y_ul=int(y_c)+1
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c), int(x_c)+1, int(x_c)+1, int(x_c)]
            y = [y_ll,y_ll,int(y_c), int(y_c)]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['correct','DOWN',x_c,y_c])
            return ['correct','DOWN',x_c,y_c]

    elif int(x_c) < X:
        x_ll=int(x_c)
        if y_c <= Y and y_c >= Y-1:
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c),x_ul,x_ul,int(x_c)]
            y = [Y-1, Y-1, Y, Y]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['RIGHT','correct',x_c,y_c])
            return ['RIGHT','correct',x_c,y_c]
        elif int(y_c) < Y:
            y_ll=int(y_c)
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c),x_ul,x_ul,int(x_c)]
            y = [int(y_c),int(y_c),y_ul,y_ul]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['RIGHT', 'UP',x_c,y_c])
            return ['RIGHT', 'UP',x_c,y_c]
        else:
            y_ul= int(y_c)+1
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c),x_ul,x_ul,int(x_c)]
            y = [int(y_c),int(y_c),y_ll,y_ll]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['RIGHT','DOWN',x_c,y_c])
            return ['RIGHT','DOWN',x_c,y_c]

    else:
        x_ul=int(x_c)+1
        if y_c <= Y and y_c >= Y-1:
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c),x_ll,x_ll,int(x_c)]
            y = [Y-1, Y-1, Y, Y]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['LEFT','correct',x_c,y_c])
            return ['LEFT','correct',x_c,y_c]
        elif int(y_c) < Y:
            y_ll=int(y_c)
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c),x_ll,x_ll,int(x_c)]
            y = [int(y_c),int(y_c),y_ul,y_ul]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['LEFT', 'UP',x_c,y_c])
            return ['LEFT', 'UP',x_c,y_c]
        else:
            y_ul= int(y_c)+1
            plt.draw()
            plt.fill(x_clear,y_clear,'white')
            plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
            x = [int(x_c),x_ll,x_ll,int(x_c)]
            y = [int(y_c),int(y_c),y_ll,y_ll]
            plt.fill(x,y,'green')
            #print ([x_ll,x_ul,y_ll,y_ul])
            #print (['LEFT','DOWN',x_c,y_c])
            return ['LEFT','DOWN',x_c,y_c]

from mpl_point_clicker import clicker
fig, ax = plt.subplots()
ax.plot(x_o,y_o)
fig.canvas.mpl_connect('button_press_event', onclick)
klicker = clicker(ax, ["click"], markers=["o"])
plt.xticks(x_o)
plt.yticks(y_o)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```
